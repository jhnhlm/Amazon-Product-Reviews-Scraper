```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 1)
```

```python
df = pd.read_csv('rev.csv')
df.head()
```

```python
#to extract rating out of "5.0 out of 5 stars" strings

def rate(text):
    x = text.split(".")[0]
    return int(x)

df['Stars'] = df['Stars'].apply(lambda x: rate(x))
```

```python
df.head()
```

```python
#
import re
def cleaner(text):
    return re.sub('\W+', ' ', text)

df['cleaned'] = df['Reviews'].apply(lambda x: cleaner(x))
df.head()
```

```python
from nltk.tokenize import word_tokenize
df['word_tok'] = df['cleaned'].apply(lambda x: word_tokenize(x))
df.head()
```

```python
#Stop words
from nltk.corpus import stopwords 
stop_words = set(stopwords.words('english'))
j = []
for i in df['word_tok']:
    d = []
    for w in i:
        if not w in stop_words:
            d.append(w)
    j.append(d)
df['no_stopword'] = j
df.head()
```

```python
#Stemming
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

j = []
for i in df['no_stopword']:
    d = []
    for w in i:
        x = stemmer.stem(w)
        d.append(x)
    j.append(d)
df['stem_word'] = j

df.head()
```
