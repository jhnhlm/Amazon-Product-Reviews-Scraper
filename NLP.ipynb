```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 1)
```

```python
df = pd.read_csv('rev.csv')
df.head()
```

```python
#to extract rating out of "5.0 out of 5 stars" strings

def rate(text):
    x = text.split(".")[0]
    return int(x)

df['Stars'] = df['Stars'].apply(lambda x: rate(x))
```

```python
df.head()
```

```python
#
import re
def cleaner(text):
    return re.sub('\W+', ' ', text)

df['cleaned'] = df['Reviews'].apply(lambda x: cleaner(x))
df.head()
```

```python
from nltk.tokenize import word_tokenize
df['word_tok'] = df['cleaned'].apply(lambda x: word_tokenize(x))
df.head()
```

```python
#Stop words
from nltk.corpus import stopwords 
stop_words = set(stopwords.words('english'))
j = []
for i in df['word_tok']:
    d = []
    for w in i:
        if not w in stop_words:
            d.append(w)
    j.append(d)
df['no_stopword'] = j
df.head()
```

```python
#Stemming
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

j = []
for i in df['no_stopword']:
    d = []
    for w in i:
        x = stemmer.stem(w)
        d.append(x)
    j.append(d)
df['stem_word'] = j

df.head()
```

```python
#Part of Speech
from nltk import pos_tag

pos_tagged_oz = list()
for i in df['word_tok']:
  pos_tagged_oz.append(pos_tag(i))
df['POS'] = pos_tagged_oz

df.head()
```

```python
from wordcloud import WordCloud
# less than 3 stars reviews 
three = df[df['Stars'] < 3]
text_list = ' '.join([' '.join(review) for review in three['no_stopword']])
wc = WordCloud(width=1000, height=500, background_color='white').generate(text_list)
wc.to_image()
```

```python
#frequency distribution
from nltk.probability import FreqDist
x = df[df.Stars < 3]
fdist = FreqDist()
for word in x['no_stopword']:
    for i in word:
        fdist[i] +=1
fdist.most_common(20)
```

```python
import nltk

flat = df['word_tok'].to_numpy().flatten().tolist()
for i in flat:
    finder = nltk.collocations.BigramCollocationFinder.from_words(i)
finder.ngram_fd.most_common(15)
```

```python
#Hash Vectorizing 
from sklearn.feature_extraction.text import HashingVectorizer

corpus = []
for i in df['cleaned']:
    corpus.append(i)

vectorizer = HashingVectorizer(n_features=10)
vector = vectorizer.transform(corpus)
m = vector.toarray()
m
```

```python
#Import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
Text = df['cleaned']
vectorizer = TfidfVectorizer()
vectorizer.fit(Text)
print(vectorizer.vocabulary_) 
print(vectorizer.idf_)
```

```python
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer

for x in df['cleaned']:
    blob = TextBlob(x, analyzer=NaiveBayesAnalyzer()) 
    for x in blob.sentiment:
        print(x)
```
